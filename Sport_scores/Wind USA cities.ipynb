```python
Jan=wind[wind['datetime'].str.contains("-01-")]
Feb=wind[wind['datetime'].str.contains("-02-")]
Mar=wind[wind['datetime'].str.contains("-03-")]
Apr=wind[wind['datetime'].str.contains("-04-")]
May=wind[wind['datetime'].str.contains("-05-")]
Jun=wind[wind['datetime'].str.contains("-06-")]
Jul=wind[wind['datetime'].str.contains("-07-")]
Aug=wind[wind['datetime'].str.contains("-08-")]
Sep=wind[wind['datetime'].str.contains("-09-")]
Oct=wind[wind['datetime'].str.contains("-10-")]
Nov=wind[wind['datetime'].str.contains("-11-")]
Dec=wind[wind['datetime'].str.contains("-12-")]
```

```python
Jan_mean=Jan.mean()
Feb_mean=Feb.mean()
Mar_mean=Mar.mean()
Apr_mean=Apr.mean()
May_mean=May.mean()
Jun_mean=Jun.mean()
Jul_mean=Jul.mean()
Aug_mean=Aug.mean()
Sep_mean=Sep.mean()
Oct_mean=Oct.mean()
Nov_mean=Nov.mean()
Dec_mean=Dec.mean()
```

```python
Janpd=pd.DataFrame(Jan_mean)
Febpd=pd.DataFrame(Feb_mean)
Marpd=pd.DataFrame(Mar_mean)
Aprpd=pd.DataFrame(Apr_mean)
Maypd=pd.DataFrame(May_mean)
Junpd=pd.DataFrame(Jun_mean)
Julpd=pd.DataFrame(Jul_mean)
Augpd=pd.DataFrame(Aug_mean)
Seppd=pd.DataFrame(Sep_mean)
Octpd=pd.DataFrame(Oct_mean)
Novpd=pd.DataFrame(Nov_mean)
Decpd=pd.DataFrame(Dec_mean)
```

```python
full_table = pd.concat([Janpd, Febpd, Marpd, Aprpd, Maypd
                       ,Junpd,Julpd,Augpd,Seppd, Octpd, Novpd,Decpd ], axis=1, sort=False)
```

```python
full_table
```

```python
full_table.columns=['Jan','Feb','Mar','Apr','May','Jun','Jul'
                   ,'Aug', 'Sep','Oct','Nov','Dec']
```

```python
full_table
```

```python
full_table.index.names = ['City']
full_table
```

```python
full_table_dest=(pd.merge(full_table, dest, on='City'))
```
